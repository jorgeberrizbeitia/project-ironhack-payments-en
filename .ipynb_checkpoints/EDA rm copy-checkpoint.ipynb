Frequency of Service Usage: Understand how often users from each cohort utilize IronHack Payments' cash advance services over time.

```python
# From cashreq.csv:
# user_id --> to group users into cohorts
# created_at_fee --> to know when the service was used
```

```python
import pandas as pd

# Load data
fees_df = pd.read_csv("project_dataset/fees.csv")
cashreq_df = pd.read_csv("project_dataset/cashreq.csv")

# Ensure created_at in fees is parsed correctly
fees_df['created_at'] = pd.to_datetime(fees_df['created_at'], errors='coerce')

# Merge to get user_id (because fees.csv doesn't include it directly)
merged_df = pd.merge(
    fees_df,
    cashreq_df[['id', 'user_id']],
    left_on='cash_request_id',
    right_on='id',
    how='left'
)

# Drop rows with missing user_id or created_at
merged_df = merged_df.dropna(subset=['user_id', 'created_at'])
merged_df['user_id'] = merged_df['user_id'].astype(str)

# STEP 1: Define cohort (user's first use of the service)
user_cohort = merged_df.groupby('user_id')['created_at'].min().reset_index()
user_cohort.rename(columns={'created_at': 'cohort_date'}, inplace=True)

# STEP 2: Merge cohort info into dataset
merged_df = pd.merge(merged_df, user_cohort, on='user_id', how='left')

# STEP 3: Extract period (month) from both dates
merged_df['cohort_month'] = merged_df['cohort_date'].dt.to_period('M')
merged_df['usage_month'] = merged_df['created_at'].dt.to_period('M')

# STEP 4: Count usage per cohort per month
usage_freq = merged_df.groupby(['cohort_month', 'usage_month'])['cash_request_id'].count().reset_index()
usage_freq.rename(columns={'cash_request_id': 'service_usage_count'}, inplace=True)

# Show result
usage_freq
```
